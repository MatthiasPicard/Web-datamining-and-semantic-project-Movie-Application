Loads the ontology and displays all the Persons (without using queries, without inference).

```python
from owlready2 import *

onto2 = get_ontology("movies_project.owl").load()
with onto2:
    people = onto2.search(is_a=onto2.myPerson)
    for p in people[1:]:
        print(p.iri,end='\n') # Display IRI
```

Loads the ontology and displays all the Persons (using a query, without inference). Create the used query in text file under the data folder.

```python
import rdflib
g = rdflib.Graph()
g.parse("movies_project.owl")
RDF = "https://schema.org/"
sparql = f""" SELECT ?p WHERE {{ ?p a  <{RDF}myPerson>}}"""
res = g.query(sparql)
for row in res:
    print(row.p) # Display IRI
```

Loads the ontology and displays all the Actors (without using queries, using inference).

```python
onto3 = get_ontology("movies_project.owl").load()
with onto3:
    sync_reasoner_pellet(infer_property_values=True, infer_data_property_values=True, debug = 2)
    actors = onto3.search(is_a=onto3.myActor)
```

```python
for a in actors[1:]:
    print(a.iri)
```

Display all persons that are actors and directors

```python
sparql = f"""SELECT ?a 
        WHERE {{
                ?a <{RDF}isActor> ?m1 .
                ?a <{RDF}isDirector> ?m2
                
                }}"""

res = g.query(sparql)
sum(1 for _ in res)
names = set()
for row in res:
    names.add(row.a)
for row in names:
    print(row)
```
